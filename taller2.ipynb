- SimpleITK: Esta biblioteca tiene como enfoque la manipulación de imágenes médicas, nos permite cargarlas y procesarlas. *Formato de archivos:  DICOM, NIfTI, Analyze, MetalImage

- VTK: Por su siglas, The Visualization ToolKit, es una biblioteca utilizada en el procesamiento y visualización de datos 3D. Nos permite la representación y animación de datos volumétricos. Formato de archivos: VTK trabaja con un formato de archivo propio, sin embargo podemos integrar otras bibliotecas para trabajar con DiCOM y NIfTI, una de estas bibliotecas es ITK, la cual procesa imágenes médicas y se encuentra bien optimizada para su uso en VTK.

- NiBabel: Esta biblioteca fue diseñada para con datos de imágenes médicas en formato NIfTI, es ampliamente utilizada en neuroimagen, puesto que las resonancias magnéticas y las tomografías por emisión de positrones son entregadas comúnmente en fomato NIfTI. Esta biblioteca puede leer imágenes en 2D Y 3D, nos proporciona acceso a los metadatos de las imágenes médicas, podemos acceder y manipular los datos con matrices de NumPy, realizar operaciones matemáticas, filtrado y segmentación.
Formato de archivos: Está optimizado para trabajar con NIfTI, pero también da soporte para Analyze, MINC y DICOM (usando dicom2nifti).

- Scikit-image: Este es un paquete que opera con matrices de Numpy, se enfoca en el procesamiento de imágenes en forma de matrices
- Formato de archivos: Numpy array, no trabaja directamente con DICOM O NIfTI, pero podemos hacer la conversión de datos a matrices de numpy y hacer el procesamiento de imágenes.
- Nilearn: Esta biblioteca nos permite visualizar, preprocesar y analizar datos de neuroimahen. Su foco principal es el análisis y procesamiento de datos de neuroimagen como las imágenes de resonancia magnética.
- Formato de archivos: NIfTI, pero también podemos trabajar con otros formatos haciendo la conversión.

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_nifti = nib.load('MRBrainTumor1.nii')
img_nifti
```

```python
#Pasar nuestra imagen a una matriz de numpy
dataimg = img_nifti.get_fdata()
```

```python
#Establecer las dimensiones de nuestra imagen
dim = dataimg.shape
plano_sagital = dataimg[:,dim[1]//2,:]
plano_axial = dataimg[dim[0]// 2,:,:]
plano_coronal = dataimg[:,:,dim[2]//2]
```

```python
#Visualización de los planos
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(np.rot90(plano_axial), cmap='gray')
plt.title('Plano Axial')

plt.subplot(132)
plt.imshow(np.rot90(plano_sagital), cmap='gray')
plt.title('Plano Sagital')

plt.subplot(133)
plt.imshow(np.rot90(plano_coronal), cmap='gray')
plt.title('Plano Coronal')
```

```python
# Acceso a los metadatos de la imagen
metadatos = img_nifti.header

# Dimensiones e información de nuestra imagen
print("Dimensiones de la imagen:", metadatos.get_data_shape())
print("Tamaño de píxel (en mm):", metadatos.get_zooms())
print("Unidades de medida:", metadatos.get_xyzt_units())
```
